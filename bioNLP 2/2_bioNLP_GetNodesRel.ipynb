```python
import csv
article_id=[]
def create_list_from_json(jsonfile):

    with open(jsonfile) as f:
        data = f.read()
    
    parsed_entities = json.loads(data)
        
    sentences=[]
    sentence_id=[]
    
    entity_id=[]
    entity_name=[]
    other_ids=[]
    entity_type=[]
    
    for s in parsed_entities:

        #Sentence: id,text
        sentences.append(s['text'].replace(',',' '))
        sentence_id.append(s['text_sha256'])
        article_id.append('1')
                
        #Entity: id,name,other_ids,type
        try:
            entities =s['entities']
            for entity in entities:
                entity_id.append(entity['entity_id'])
                entity_name.append(entity['entity_id'])
                other_ids.append(entity['other_ids'])
                entity_type.append(entity['entity_type'])
        except:
            pass

    
    data={'id':sentence_id,'text':sentences}
    df = pd.DataFrame(data)
    df.to_csv('sentences.csv')
    
    data={'sid':article_id,'did':sentence_id}
    df = pd.DataFrame(data)
    df.to_csv('article_sentences.csv')

    data={'id':entity_id,'name':entity_id,'other_ids':other_ids,'type':entity_type}
    df = pd.DataFrame(data)
    df.to_csv('entity.csv')


create_list_from_json('parsed_entities.json')  # create the row to be added to csv for each file (json-file)

with open('predicted_rels.json') as f:
    predicted_rels = f.read()
    
predicted_rels = json.loads(predicted_rels)
print(predicted_rels)

df = pd.DataFrame(predicted_rels)
df.to_csv('relationship.csv')
```
